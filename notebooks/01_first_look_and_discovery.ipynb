## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

### Custom Discovery: Correlation for Active Pages
Does `search_volume` predict traffic better if we exclude pages that received zero impressions over the last 90 days?

```python
active_pages = df[df['impressions_90d'] > 0]
active_corr = active_pages['search_volume'].corr(active_pages['impressions_90d'])

print(f"Correlation for all pages: {corr:.3f}")
print(f"Correlation for active pages (impressions > 0): {active_corr:.3f}")

if abs(active_corr) < 0.1:
    print("\nObservation: Even for active pages, the correlation remains near zero. Search volume is a poor predictor of actual impressions.")
```

### Custom Discovery: CTR by Content Type
Does the type of content affect the click-through rate, even when pages are at the same position tier (e.g., 'top_3')?

```python
# Let's look at the 'top_3' tier specifically
top_3_data = df[(df['position_tier'] == 'top_3') & (df['impressions_90d'] >= 100)]
ctr_by_type = top_3_data.groupby('content_type')['ctr'].mean().sort_values()

print("Mean CTR by Content Type for Top 3 positions:")
print(ctr_by_type.round(4).to_string())

print(f"\nObservation: The '{ctr_by_type.index[0]}' type has the worst CTR in the top 3 tier.")
```

### Custom Discovery: Avg Position vs CTR
Let's test the belief that better average positions strictly correlate with higher CTRs by plotting them against each other.

```python
import matplotlib.pyplot as plt

# Filter for pages with enough data to have a meaningful CTR
active_visible = df[(df['impressions_90d'] >= 500) & (df['avg_position'] <= 20)]

plt.figure(figsize=(10, 6))
plt.scatter(active_visible['avg_position'], active_visible['ctr'], alpha=0.3)
plt.title('Average Position vs. CTR (min 500 impressions)')
plt.xlabel('Average Position')
plt.ylabel('CTR')
plt.gca().invert_xaxis() # Better position is on the left
plt.show()

pos_ctr_corr = active_visible['avg_position'].corr(active_visible['ctr'])
print(f"Correlation between Avg Position and CTR: {pos_ctr_corr:.3f}")
print("\nObservation: There is a clear directional trend where better (lower) positions correlate with higher CTR.")
```
